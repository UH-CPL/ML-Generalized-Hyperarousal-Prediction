```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

import seaborn as sns

import cv2
from PIL import Image as img


from sklearn.metrics import accuracy_score, f1_score, recall_score, precision_score, confusion_matrix, ConfusionMatrixDisplay




data_path = '../../data/sim1/'
figure_path = '../../figure/sim1/'
```

```python
otsu_df = pd.read_csv(data_path + 'otsu_data.csv')
otsu_df.head()
```

----------------------------

#### Otsu's thresholding method involves iterating through all the possible threshold values and calculating a measure of spread for the value levels each side of the threshold.

#### For bimodal distribution, Otsu's algorithm tries to find a threshold value (t) which 
- minimizes "within-class variance"
- maximizes "between class variance" of the segmented classes

#### Optimal thresholding - Finding the optimal threshold(s) for a multimodal histogram is described in well-known literature as a problem in fitting a sum of Gaussians to the histogram. This fitting problem is shown experimentally to be a nonlinear minimization with local minima.

----------------------------

```python
mult_val = 255

def get_otsu_threshold(pred_df, label):
        data = np.reshape(list(pred_df[label]*mult_val), [1, -1])
        # print(data)
        
        #################################################
        #                  Bad Coding
        #################################################
        image = img.fromarray(data)
        image.convert("L").save('temp_pic.png')
        image = cv2.imread('temp_pic.png', 0)
        #################################################

        # cv2.THRESH_BINARY+cv2.THRESH_OTSU
        thr, _ = cv2.threshold(image, 0, 1, cv2.THRESH_BINARY+cv2.THRESH_OTSU)

        return thr/mult_val
    
    
def set_prediction(row, label, otsu_threshold):
    if row[label] < otsu_threshold:
        return 'stressed' if label=='relaxed' else 'relaxed'
    else:
        return label
    
    
    
def get_confusion_matrix(signal_df, arousal_signal, label, y_test, y_pred, otsu_threshold, do_normalize):
    
    #####################################################################################
    fig = plt.figure(figsize=(10, 10))
    ax = fig.add_subplot(111)

    sns.set(font_scale=1.4)
    
    #-------------------------------------------------------------#
    # labels = ["normal", "relaxed", "stressed", "unknown"]
    # labels = ["relaxed", "normal", "unknown", "stressed"]
    # labels = ["relaxed", "stressed"]
    labels = sorted(signal_df.Arousal_Mode.unique())
    #-------------------------------------------------------------#
    
    if do_normalize:
        conf_mat = pd.DataFrame(confusion_matrix(y_test, y_pred, labels=labels, normalize='all'))
        sns.heatmap(conf_mat*100, annot=True, annot_kws={"size": 16})
        plt.title('Confusion Matrix (%) - ' + arousal_signal + ' - ' + label + ' - ' + str(round(otsu_threshold, 2)))
        plot_name = arousal_signal.lower() + '_' + label + '_percentage'
        
    else:
        conf_mat = pd.DataFrame(confusion_matrix(y_test, y_pred, labels=labels))
        sns.heatmap(conf_mat, annot=True, fmt='d', annot_kws={"size": 16})
        plt.title('Confusion Matrix - ' + arousal_signal + ' - ' + label + ' - ' + str(round(otsu_threshold, 2)))
        plot_name = arousal_signal.lower() + '_' + label

    ax.set_xticklabels(labels)
    ax.set_yticklabels(labels)

    plt.xlabel('Predicted')
    plt.ylabel('Actual')

    # plt.show()

    plt.savefig(figure_path + plot_name + '_confusion_matrix.png')
    plt.savefig(figure_path + plot_name + '_confusion_matrix.pdf')
    #####################################################################################
    
    
```

```python
legend_dict = { 'Relaxed' : 'lightgreen', 'Stressed' : 'red' }
patches = [mpatches.Patch(color=legend_dict[key], label=key) for key in legend_dict]

def get_probability_distribution(arousal_signal):
    signal_df = pd.read_csv('../../data/sim1/' + arousal_signal.lower() + '_pred_result_df.csv')
    signal_df.rename(columns={'Relaxed_Prob': 'relaxed', 'Stress_Prob': 'stressed'}, inplace=True)

    for label in sorted(signal_df.Arousal_Mode.unique()):
        pred_df = signal_df.copy()[signal_df.Arousal_Mode==label]
        
        plt.figure(figsize = (22, 10))
        plt.title(arousal_signal + ' - Ground Truth - ' + label.capitalize(), fontsize=64)
        plt.xlabel('Probability of the label - ' + label, fontsize=28)  
        
        
        otsu_threshold = get_otsu_threshold(pred_df, label)
        # print(label, otsu_threshold)
        
    
####################################################################################################    
#         sns.kdeplot(data=pred_df,
#                     x=label,
#                     hue="Prediction", 
#                     multiple="stack",
#                     palette="bright",
#                     fill=True,
#                     alpha = 0.3)
    
#         otsu_threshold = get_otsu_threshold(pred_df, label)
#         # otsu_threshold = get_otsu_threshold(pred_df, label) # From R method
        
#         plt.axvline(otsu_threshold, 0, 999, color = 'red')
    
#         plt.ylabel('Density', fontsize=28)
        
#         plt.savefig(figure_path + arousal_signal.lower() + '_' + label + '_prob_dist.png')
#         plt.savefig(figure_path + arousal_signal.lower() + '_' + label + '_prob_dist.pdf')
####################################################################################################

####################################################################################################
        sns.histplot(data=pred_df,
                    x=label,
                    hue="Prediction", 
                    ## multiple="stack",
                    kde=True,
                    palette="bright",
                    fill=True,
                    alpha = 0.3,
                    bins = 25)

        plt.ylabel('Frequency', fontsize=28)
    
        plt.savefig(figure_path + arousal_signal.lower() + '_' + label + '_prob_dist.png')
        plt.savefig(figure_path + arousal_signal.lower() + '_' + label + '_prob_dist.pdf')
####################################################################################################
        
        y_test = signal_df.Arousal_Mode
        y_pred = signal_df.apply(lambda row: set_prediction(row, label, otsu_threshold), axis=1)
        print('\n------ '+ label + '------')
        print('Accuracy: ' + str(round(accuracy_score(y_test, y_pred), 2)))
        print('F1: ' + str(round(f1_score(y_test, y_pred, average='weighted'), 2)))
        print('Recall: ' + str(round(recall_score(y_test, y_pred, average='weighted'), 2)))
        print('Precision: ' + str(round(precision_score(y_test, y_pred, average='weighted'), 2)))
        
        get_confusion_matrix(signal_df, arousal_signal, label, y_test, y_pred, otsu_threshold, True)
 
```

```python
#####################################################################################
# arousal_signals = ['PP']
arousal_signals = ['PP', 'PP_2', 'HR', 'BR', 'PP_HR_BR']

# 'PP', 'PP_2', 'HR', 'BR'
# 'PP_HR', 'HR_BR', 'PP_BR', 'PP_HR_BR'
#####################################################################################


for arousal_signal in arousal_signals:
    print('----------------------------------------------------')
    print('------------------------', arousal_signal, '------------------------')
    print('----------------------------------------------------\n')
        
    get_probability_distribution(arousal_signal)
        
    print('\n\n')
```
